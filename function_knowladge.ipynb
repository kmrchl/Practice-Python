```python
print("Write a Python function to convert temperature")
print("")

def convert_temperature(temperature, scale):
    """
    Converts a temperature value from one scale to another.

    Parameters:
        temperature : The temperature value to be converted.
        scale       : The scale of the input temperature. 'C' for Celsius, 'F' for Fahrenheit.

    Returns:
        The converted temperature value.

    Raises:
        ValueError: If an invalid scale is provided.

    Examples:
        >>> convert_temperature(32, 'F')
        0.0
        >>> convert_temperature(100, 'C')
        212.0
    """
    if scale not in ['C', 'F']:
        raise ValueError("Invalid scale. Use 'C' for Celsius or 'F' for Fahrenheit.")
    
    if scale == 'C':
        converted_temperature = (temperature * 9/5) + 32
    else:
        converted_temperature = (temperature - 32) * 5/9
    
    return converted_temperature

result1 = convert_temperature(32, 'F')
print(f"32 degrees Fahrenheit is {result1} degrees Celsius.")

# Example 2: Convert 100 degrees Celsius to Fahrenheit
result2 = convert_temperature(100, 'C')
print(f"100 degrees Celsius is {result2} degrees Fahrenheit.")

print("<=============================================>")
print("")
print("Create a calculator")
print("")

def arithmetic_calculator(operator, *args):
    """
    Performs arithmetic operations on a variable number of arguments.

    Parameters:
        operator : The arithmetic operator to be applied. Supported operators: '+', '-', '*', '/'.
        *args    : Variable number of arguments.

    Returns:
        The result of the arithmetic operation.

    Raises:
        ValueError: If an invalid operator is provided.
        ZeroDivisionError: If division by zero occurs.

    Examples:
        >>> arithmetic_calculator('+', 2, 3, 4)
        9
        >>> arithmetic_calculator('*', 1, 2, 3, 4)
        24
    """
    if operator not in ['+', '-', '*', '/']:
        raise ValueError("Invalid operator. Use '+', '-', '*', or '/'.")

    if len(args) < 2:
        raise ValueError("At least two arguments are required for arithmetic operations.")
    
    result = args[0]

    if operator == '+':
        for num in args[1:]:
            result += num
    elif operator == '-':
        for num in args[1:]:
            result -= num
    elif operator == '*':
        for num in args[1:]:
            result *= num
    elif operator == '/':
        for num in args[1:]:
            if num == 0:
                raise ZeroDivisionError("Division by zero is not allowed.")
            result /= num

    return result

result1 = arithmetic_calculator('+', 2, 3, 4)
print(f"2 + 3 + 4 = {result1}")

# Example 2: Multiplication
result2 = arithmetic_calculator('*', 1, 2, 3, 4)
print(f"1 * 2 * 3 * 4 = {result2}")
```
