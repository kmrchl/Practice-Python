```python
print("Praktisi Python")
print("<======================================>")
print("")

print("Argument - Positional")
def greet(name, greeting):
    print(f"{greeting}, {name}!")
greet("Alice", "Hello") # "Hello, Alice!"

print("------------------------------------")

print("Argument - Keyword")
def greet(name, greeting):
    print(f"{greeting}, {name}!")
greet(greeting="Hi", name="Bob")

print("------------------------------------")

print("Argument - Default")
def greet(name, greeting="Hello"):
    print(f"{greeting}, {name}!")
greet("Eve")

print("------------------------------------")

print("Argument Arbitary Args")

def print_args(*args):
    for arg in args:
        print(arg)
print_args(1, "apple", True) 

print("------------------------------------")

print("Argument Arbitary kwargs")

def print_kwargs(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")
print_kwargs(name="Alice", age=30)

print("------------------------------------")

print("Arguments - Passing Lists/Dictionaries/etc")

def print_list(numbers):
    for num in numbers:
        print(num)
print_list([1, 2, 3, 4]) # 1\n2\n3\n4
print("")
def print_dict(my_dict):
    for key, value in my_dict.items():
        print(f"{key}: {value}")
print_dict({"name": "Bob", "age": 25})

print("------------------------------------")

print("Different argument combinations")
def f(a, *b, c=6, **d): # compare it with def f(a, b=6, *c, **d)
    print(f"a: {a}")
    print(f"b: {b}")
    print(f"c: {c}")
    print(f"d: {d}")
f(1, 2, 3, x=4, y=5) # Default used
f(1, 2, 3, c=7, x=4, y=5)

print("------------------------------------")

print("Return Statements - Examples")

def add_numbers(a, b):
    """This function adds two numbers and returns the result."""
    result = a + b
    return result
# Call the function with a return value
sum_result = add_numbers(5, 3)
print(f"The sum is: {sum_result}")

print("")

def greet(name):
    """This function greets the person passed in as a parameter."""
    print(f"Hello, {name}!")
# Call the function without a return value
greet("Alice")

print("------------------------------------")

print("It is executed at runtime")

if 'a' == 'a':
    def greet():
        return "Hello, World!"
else:
    def greet():
        return "Hi there!"

print("")

def say_hello():
    print("Hello, World!")
greeting = say_hello # Assigning the function to a different name

greeting() # Calls the function
def say_hello():
    print("Hello, Python!")
say_hello() # Calls the updated function

print("------------------------------------")

print("It is first class object")

def greet(name):
    return f"Hello, {name}!"
# Assign the function to a variable
my_function = greet
# Call the function using the variable
result = my_function("Alice")
print(result) # Output: Hello, Alice!

print("------------------------------------")

print("Scope")

def my_function():
    x = 10 # x is in the local scope
    print(x)
def outer_function():
    y = 20 # y is in the enclosing scope
    def inner_function():
        print(y) # inner_function can access y from the enclosing scope
z = 30 # z is in the global scope
def another_function():
    print(z) # another_function can access z from the global scope
print(len("Hello, World!")) # len is a built-in function


print("------------------------------------")

print("Scope - Global keyword")

global_var = 10 # Global variable
def modify_global():
    global global_var # Use the global keyword to MODIFY the global variable
    global_var = 20
modify_global()
print(global_var)

print("------------------------------------")

print("Pass by Object Reference")

def modify_list(my_list):
    my_list.append(4) # Modifying the list inside the function
my_list = [1, 2, 3]
modify_list(my_list)
print(my_list) # Output: [1, 2, 3, 4]

print("")

def reassign_list(my_list):
    my_list = [4, 5, 6] # Reassigning the list to a new object
my_list = [1, 2, 3]
reassign_list(my_list)
print(my_list) # Output: [1, 2, 3]

print("------------------------------------")

print("Function Overloading")

def product(a, b):
    print(a * b)
def product(a, b, c):
    print(a * b * c)
# product(4, 5) # Uncommenting this shows an error

product(4, 5, 5) # This line will call the last function
# Try using *args to solve that problem

print("------------------------------------")

print("Function Recursion")
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)

result = factorial(5)
print(result) # Output: 120

print("------------------------------------")

print("Docstrings")

def add(a, b):
    """ This function adds two numbers together.
    Args:
        a (int): The first number to be added.
        b (int): The second number to be added.

    Returns:
        int: The sum of the two input numbers.

    Example:
        >>> add(3, 4)
        7
    """
    return a + b
help(add) 
```
